### Data frames <a class="anchor" id="data-frames"></a>
The python package that is most commonly used to work with spreadsheet data is called `pandas`, the name is derived from "panel data", an econometrics term for multidimensional structured data sets. Spreadsheet data are easily loaded into pandas from `.csv` or other spreadsheet formats. I do not have any good data set lying around, so I will load a public dataset from the web (you can view the data by pasting the url into your browser). I store my dataset in a variable called `iris`, that I can use to access the data from now on.

```python
import pandas as pd

iris = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
```

To have a quick peak at the data, I can type `iris.head(<number_of_rows>)`. Notice that I do not have to use the `pd.`-syntax for this. `iris` is now a pandas dataframe, and all pandas data frames have a number of inbuilt functions (called methods) that can be used directly from the dataframe instead of by calling `pandas` separately.

```python
iris.head(10)
```

As you can see, Python starts numbering the rows at 0, which is confusing at first, but a standard in many programming languages and you get used to it quickly. I can easily find out how many rows and columns my data set has.

```python
iris.shape
```

If you want to be able to see all the rows at a time, you can either display them all in this notebook, or just have your excel spreadsheet open next to this notebook. You can easily get a basic description of the data in the different columns.

```python
iris.describe()
```

Instead of looking at the general description of the data, we might want to make more specific comparisons, such as between the mean values of the different species.

```python
# We will reuse this data frame, so I save it as a variable 
species_comparison = iris.groupby('species').mean()
species_comparison # This is just to make the table appear below
```

We can see that there are clear differences between species, but they might be even clearer if we display them graphically in a chart.

### Plots <a class="anchor" id="Plots"></a>

```python
# Prevent plots from popping up in a new window
%matplotlib inline

species_comparison.plot(kind='bar')
```

Depending in what your are interesting in showing, it could be useful to have the species as the different colors and the columns along the xaxis. We can easily achieve this by transposing (`.T`) our data frame.

```python
species_comparison.T.plot(kind='bar')
```

The graph looks pretty good already, but we can customized its appearance if we want to. Instead of changing each bar and line one at a time, we can easily apply a new style to entire graph. To change the style, we need to use the specific plotting package `matplotlib`, which is also what pandas uses for its plots. `matplotlib` allows us to [define a stylesheet](http://matplotlib.org/users/style_sheets.html) for our plot. I will use one called `ggplot`, that is inspired by an R-package with the same name.

```python
import matplotlib.pyplot as plt

# The `with`-keyword is used to apply this style only to the immediately following indented code,
# instead of all plots in the notebook
with plt.style.context('ggplot'):
     # I will also add the standard deviation as the y error bars
    species_comparison.T.plot(kind='bar',
        yerr=iris.groupby('species').std().T)
```

Another plotting library `seaborn`, which builds upon `matplotlib`, by adding new styles, additional plot types and some easily accessible statistical measures.

```python
import seaborn as sns

sns.barplot(y='sepal_length', x='species', data=iris)
```

To get the exact same plot as we had previously, we first need to reshape our data frame into the [tidy or molten data format](http://vita.had.co.nz/papers/tidy-data.pdf), which is the format `seaborn` prefers. This is sometimes refered to as changing the data from wide to tall or long format, and moves some of the information from columns to rows:

```python
Image('./molten-data.png') # Image from http://vita.had.co.nz/papers/tidy-data.pdf
```

We can use pandas built-in `melt`-function to transform the data into the molten format.

```python
iris_melt = pd.melt(iris, id_vars='species')
iris_melt.head()
```

Now we can recreate the plot we made earlier. We do not need to call `groupby` or `mean` on the `iris`-dataframe when plotting with `seaborn`. Instead we control these options from `seaborn` with the plot type we chose (barplot = mean automatically) and the `hue`-parameter, which is analogous to `groupby`.

```python
sns.barplot(x='variable', y='value', hue='species', data=iris_melt)
```

`seaborn` also makes it easy to change the deafult style.

```python
sns.set(style='white', context='talk')
sns.barplot(x='variable', y='value', hue='species', data=iris_melt)
```

 Now that we know how to use seaborn, we can create more insigthful plots with less effort.

```python
sns.swarmplot(x='variable', y='value', hue='species', data=iris_melt, split=True, palette='Set2')
sns.despine() # removes the right and top black border
```

```python
with sns.plotting_context('talk') and sns.axes_style("ticks"):
    sns.boxplot(y='variable', x='value', hue='species', data=iris_melt, color='c', )
    sns.stripplot(y='variable', x='value', hue='species', data=iris_melt, size=2.5, palette=['k']*3, jitter=True, split=True)
sns.despine(trim=True)
```

It is also possible to get too fancy and accidentally hide important messages in the data. However, the fact that you now have access to several ways to plot your data forces you to consider what is actually important and how you can best communicate that message, rather than always making the same plot without considering its strengths and weaknesses.
